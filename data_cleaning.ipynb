```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('startup_funding.csv')
df.info()
```

```python
# Drop Remarks Column
df.drop(columns=['Remarks'],inplace=True)

# Set Sl No Column as index
df.set_index('Sr No',inplace=True)

# Renaming Column Names
df.rename(columns={'Date dd/mm/yyyy': 'Date',
                  'Startup Name': 'Startup',
                  'Industry Vertical': 'Vertical',
                  'City  Location': 'City',
                  'Investors Name': 'Investor',
                  'InvestmentnType': 'Round',
                  'Amount in USD': 'Amount'},inplace=True)
```

```python
df.sample(3)
```

## Preprocessing Amount Column

```python
df['Amount']=df['Amount'].str.replace(',','')
df['Amount']=df['Amount'].str.replace('undisclosed','0')
df['Amount']=df['Amount'].str.replace('unknown','0')
df['Amount']=df['Amount'].str.replace('Undisclosed','0')
df['Amount']=df['Amount'].str.replace('\\\\xc2\\\\xa0','')
df['Amount']=df['Amount'].str.replace('\\\\xc2\\\\','')
df['Amount']=df['Amount'].str.replace('\\\\xa0','')
df['Amount']=df['Amount'].str.replace('\\','')
df['Amount']=df['Amount'].str.replace('+','')
df['Amount']=df['Amount'].str.replace('N/A','0')
df['Amount'].fillna('0',inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('float')
```

```python
df['Amount'] = np.round(82.5*df['Amount']/10000000,2)
```

```python
df.head(5)
```

## Preprocessing Date Column

```python
print(df['Date'].unique())
```

```python
import re

# Function to check if a string matches the format xx/xx/xxxx
def is_valid_date_format(date_string):
    pattern = r'^\d{2}/\d{2}/\d{4}$'
    return re.match(pattern, date_string) is not None

# Filter rows that don't match the desired format
# invalid_dates = df[~df['date_column'].apply(is_valid_date_format)]
```

```python
df[~df['Date'].apply(is_valid_date_format)]['Date'].unique()
```

```python
df['Date'] = df['Date'].str.replace('05/072018','05/07/2018')
df['Date'] = df['Date'].str.replace('01/07/015','01/07/2015')
df['Date'] = df['Date'].str.replace('\\\\xc2\\\\xa010/7/2015','10/07/2015',regex = False)
df['Date'] = df['Date'].str.replace('12/05.2015','12/05/2015')
df['Date'] = df['Date'].str.replace('12/05.2015','12/05/2015')
df['Date'] = df['Date'].str.replace('13/04.2015','13/04/2015')
df['Date'] = df['Date'].str.replace('15/01.2015','15/01/2015')
df['Date'] = df['Date'].str.replace('22/01//2015','22/01/2015')
df['Date'] = df['Date'].str.replace('15-01-07 00:00:00','15/01/2007')
```

```python
df[~df['Date'].apply(is_valid_date_format)]['Date'].unique()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```
